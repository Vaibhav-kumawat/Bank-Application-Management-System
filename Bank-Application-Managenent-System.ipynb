```python
data = {
        "accno" : [1001, 1002, 1003, 1004, 1005],
        "name" : ['vaibhav', 'abhinav', 'aman', 'ashish', 'pramod'],
        "balance" : [10000, 12000, 7000, 9000, 10000],
        "password" : ['admin', 'adminadmin', 'passwd', '1234567', 'amigo'],
        "security_ques" : ['2101', '3103', '2004', '1910', '1208']
        }
```

```python
from getpass import getpass
import time

def password_validation(passwd):
    '''Check password is valid or not'''
    n = len(passwd)
    if n>=8:
        s = 0; u = 0; l = 0; d = 0
        for i in range(n):
            ch = passwd[i]
            if ch in "1234567890":
                d += 1
            elif ch.isupper():
                u += 1
            elif ch.islower():
                l += 1
            elif ch in "!@#$%^&*()-+":
                s += 1
            i += 1
        if u>=1 and l>=1 and s>=1 and d>=1:
            time.sleep(3)
            return True ## if your password is valid and strong...
        else:
            time.sleep(3)
            return False ## if your password not valid 
    else:
        time.sleep(3)
        return False ## if your password is not valid
    
def verify_Security_ans(question):
    '''Check security answer is valid or not'''
    n = len(question)
    if n == 4:
        dd = int(question[:2])
        mm = int(question[2:])
        if mm == 2:
            if 1<=dd<=29:
                time.sleep(3)
                return True
            else:
                time.sleep(3)
                return False
        elif mm in (1,3,5,7,8,10,12):
            if 1<=dd<=31:
                time.sleep(3)
                return True
            else:
                time.sleep(3)
                return False
        elif mm in (4,6,9,11):
            if 1<=dd<=30:
                time.sleep(3)
                return True
            else:
                time.sleep(3)
                return False
        else:
            time.sleep(3)
            return False
    time.sleep(3)
    return False

def credit_amount(index):
    user_index = index
    amount = int(input("Enter amount to credit : "))
    print(amount)
    new_amount = data["balance"][user_index] + amount
    data["balance"].remove(data["balance"][user_index])
    data["balance"].insert(user_index,new_amount)
    print("Updated amount after credit -->",data["balance"][user_index])
    time.sleep(3)
    
def debit_amount(index):
    user_index = index
    amount = int(input("Enter amount to debit : "))
    print(amount)
    if amount<=data["balance"][user_index]:
        new_amount = data["balance"][user_index] - amount
        data["balance"].remove(data["balance"][user_index])
        data["balance"].insert(user_index,new_amount)
        print("Updated amount after debited -->",data["balance"][user_index])
        time.sleep(3)
    else:
        print("Insufficient balance in your account")
        time.sleep(3)

def change_password(index):
    user_index = index
    old_passwd = getpass("Enter old password : ")
    if old_passwd == data["password"][user_index]:
        new_password = getpass("Enter new password : ")
        con_password = getpass("Re-enter new password : ")
        if new_password == con_password:
            if password_validation(new_password):
                print("Your password is valid and strong...")
                data["password"].remove(data["password"][user_index])
                data["password"].insert(user_index,new_password)
                print("Password has been changed")
                time.sleep(3)
            else:
                print("Your password is not valid")
                time.sleep(3)
        else:
            print("Passwords don't match \nPlease try again")
            time.sleep(3)
    else:
        print("Password dosen't match our records \nPlease enter correct password")
        time.sleep(3)

def login(index):
    index = index
    for n in range(10):
        print("-"*40)
        print("\n 1. credit_amount \n 2. debit_amount \n 3. change_password \n 4. check_details \n 5. logout")
        print("-"*40)
        perform = int(input("Choose any option from above : "))
        if perform == 1: 
            credit_amount(index)
            continue
        elif perform == 2:
            debit_amount(index)                    
            continue
        elif perform == 3:
            change_password(index)                    
            continue
        elif perform == 4:
            bal = data['balance'][index]
            name = data['name'][index]
            accno = data['accno'][index]
            print("DETAILS".center(50))
            print(f"NAME : {name}".center(50, "_"))
            print(f"ACCOUNT_NUMBER : {accno}".center(50, "_"))
            print(f"BALANCE : {bal}".center(50, "_"))
            time.sleep(3)
            continue
        elif perform == 5:
            print("Logout")
            print("-"*40)
            time.sleep(3)
            break
        else:
            print("Please choose correct option")
            continue

def signup():
    user_name = input("Enter user name to open account : ")
    initial_balance = int(input("Submit some ammount if you want : "))
    for k in range(3):
        password = getpass("Enter password to security : ")
        con_password = getpass("Re-enter password : ")
        if password == con_password:
            if password_validation(password):
                for i in range(3):
                    security_ans = input("Enter a security answer in ddmm : ")
                    if verify_Security_ans(security_ans):
                        data["name"].append(user_name)
                        index = data["name"].index(user_name)
                        data["balance"].insert(index,initial_balance)
                        data["password"].insert(index,password)
                        account_number = data["accno"][len(data["accno"]) - 1] + 1
                        data["accno"].insert(index,account_number)
                        data["security_ques"].insert(index,security_ans)
                        bal = data['balance'][index]
                        name = data['name'][index]
                        accno = data['accno'][index]
                        print("DETAILS".center(50))
                        print(f"NAME : {name}".center(50, "_"))
                        print(f"ACCOUNT_NUMBER : {accno}".center(50, "_"))
                        print(f"BALANCE : {bal}".center(50, "_"))
                        break
                    else:
                        print("Enter valid security answer")
                        continue
            else:
                print("Invalid password \nPlease try again")
                continue
        else:
            print("Passwords don't match \nPlease try again")
            continue
        break
    time.sleep(3)

def forgot_password():
    for i in range(1,4):
        account = int(input("Enter your account number : ")) ## Take account number for reset password
        if account in data["accno"]:
            index = data["accno"].index(account)
            verify = input("Enter security question : ") ## Take security question for verification
            if verify == data["security_ques"][index]:
                for k in range(3):
                    New_password = getpass("Enter new password : ")
                    Con_password = getpass("Re-enter new password : ")
                    if New_password == Con_password:
                        if password_validation(New_password):
                            print("Your password is valid and strong...")
                            print("Password changed")
                            data["password"].remove(data["password"][index])
                            data["password"].insert(index,New_password)
                            break
                        else:
                            print("Your password is not valid \nPleas try again")
                            continue
                    else:
                        print("Passwords don't match \nPlease try again")
                        continue
            else:
                print("Your security question is wrong \nPlease try again")
                continue
        else:
            print("This account is not exist \nPlease enter valid account number")
            continue
        break
    time.sleep(3)

def main():
    print("-"*80)
    print("BANK APPLICATION".center(80))
    print("-"*80)
    for j in range(1,11):
        print("\n 1. Login \n 2. Signup \n 3. Forgot_password \n 4. Exit")
        print("-"*40)
        ch_opt = int(input("Choose any option from above : "))
        if ch_opt == 1:
            print("-"*40)
            for i in range(1,4):
                user_accno = int(input("Enter account number : "))
                if user_accno in data["accno"]:
                    user_index = data["accno"].index(user_accno)
                    user_passwd = getpass("Enter password : ")
                    if user_passwd == data["password"][user_index]:
                        login(user_index)
                        break
                    else:
                        print("Password is incorrect \n Please enter correct password")
                        continue
                else:
                    print("This account is not in our database")
                    continue
            continue
        elif ch_opt == 2:
            print("-"*40)
            print("Creat Account")
            print("-"*15)
            signup()
            print("\nSignout")
            print("-"*40)
            continue
        elif ch_opt == 3:
            print("-"*40)
            print("Forgot password \nCreate new password")
            print("-"*20)
            forgot_password()
            print("-"*40)
            continue
        elif ch_opt == 4:
            print("Exiting")
            time.sleep(3)
            print("_"*80)
            break
        else:
            print("Please choose correct option")
            continue
```
